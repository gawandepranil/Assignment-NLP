```python
import nltk
from nltk.tokenize import (
    word_tokenize,
    wordpunct_tokenize,
    TreebankWordTokenizer,
    TweetTokenizer,
    MWETokenizer
)
from nltk.stem import PorterStemmer, SnowballStemmer
from nltk.stem import WordNetLemmatizer
```

```python
nltk.download('punkt')
nltk.download('wordnet')
nltk.download('omw-1.4')
```

```python
text = "NLTK is amazing! I'm learning tokenization, stemming & lemmatization ðŸ˜Š #NLP"
```

```python
whitespace_tokens = text.split()
print("Whitespace Tokenization:", whitespace_tokens)
```

```python
punct_tokens = wordpunct_tokenize(text)
print("Punctuation-based Tokenization:", punct_tokens)
```

```python
treebank = TreebankWordTokenizer()
treebank_tokens = treebank.tokenize(text)
print("Treebank Tokenization:", treebank_tokens)
```

```python
tweet_tokenizer = TweetTokenizer()
tweet_tokens = tweet_tokenizer.tokenize(text)
print("Tweet Tokenization:", tweet_tokens)
```

```python
mwe = MWETokenizer([('machine', 'learning'), ('natural', 'language', 'processing')])
mwe_text = "I love machine learning and natural language processing"
mwe_tokens = mwe.tokenize(mwe_text.split())
print("MWE Tokenization:", mwe_tokens)
```

```python
porter = PorterStemmer()
porter_stems = [porter.stem(word) for word in treebank_tokens]
print("Porter Stemmer:", porter_stems)
```

```python
snowball = SnowballStemmer("english")
snowball_stems = [snowball.stem(word) for word in treebank_tokens]
print("Snowball Stemmer:", snowball_stems)
```

```python
lemmatizer = WordNetLemmatizer()
lemmatized_words = [lemmatizer.lemmatize(word) for word in treebank_tokens]
print("Lemmatization:", lemmatized_words)
```
